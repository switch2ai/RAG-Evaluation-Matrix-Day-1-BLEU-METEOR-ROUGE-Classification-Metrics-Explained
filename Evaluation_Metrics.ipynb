###Classification Metrics :

```python
## For evaluation : labelled data

------------ Actual value ---
This is good movie                   Positive
This is bad movie                    Negative

----------- Predicted value --------> Predicted by model
This is good movie                   --
This is bad movie                    --
```

```python
## Confusion Metrics :

                        Predicted Positive                Predicted negative
Actual positive              TP                                  FN

Actual Negative              FP                                  TN
```

```python
## Accuracy : How many samples are correctly classified
Accuracy = (TP + TN) / TP + TN + FP + FN

## Can not trust accuracy when data is imbalance
```

```python
## Precision : What proportional of total predicted positives are actual positives
Precision = TP / (TP + FP)
```

```python
## Recall : what proportional of actual positive are correctly predicted
Recall = TP / (TP + FN)
```

```python
## F1-Score : Harmonic mean of precision & Recall
F1-Score = 2 * Precision * Recall / (Precision + Recall)
```

```python
# False Positive : ex: spam detection   --> Precision   (minimize FP)
# False Negative : ex : cancer detection --> Recall  (minimize FN)
```

### BLEU (Bilingual Evaluation Understudy)

```python
## BLEU : It is Precision based matrix
```

```python
## Precision : What proportional of total predicted positives are actual positives
Precision = TP / (TP + FP)

Reference : The cat is on mat     ---> Actual
Predicted : There is cat on mat   ---> Predicted by model

>> Precision in translation : How many words predicted by model are actually in reference

Precision = m / w_translated
m = no of matching words in a reference & predicted
w_translated = no. of words predicted
```

```python
# precision = 0 + 1 + 1 + 1 + 1 / 5 = 4/5
```

```python
Reference : The cat is on mat
Predicted : There is cat cat cat

# precision = 0 + 1 + 1 + 1 + 1 / 5 = 4/5

Modified precision :
Clipping :
If any word occuring multiple times in generated text,
only consider that word, how many times it is occuring in reference

# modified precision = 0 + 1 + 1 + 0 + 0 / 5 = 2/5
```

```python
reference : cat chased dog
Predicted : dog chased cat

referece unigram = [cat, chased, dog]
predicted unigram = [dog, chased, cat]

referece bigram = [cat chased, chased dog]
predicted bigram = [dog chased, chased cat]

referece trigram = [cat chased dog]
predicted trigram = [dog chased cat]

## N-gram Precision : unigram precision * bigram precision * trigram precision ...
```

```python
Reference : The cat is on mat
Predicted : The cat

reference unigram = [The, cat, is, on, mat]
predicted unigram = [The, cat]

reference bigram = [The cat, cat is, is on, on mat]
predicted bigram = [The cat]

unigram precision = 1 + 1 / 2 = 1
bigram precision = 1/1        = 1

# N-gram Precision = unigram precision * bigram precision
```

```python
Reference : The cat is on mat
Predicted : The cat

# Brevity Penalty :
BP =
if len(predicted) >=len (reference):
  BP = 1
else:
  BP = exp(1-len(predicted)/len(reference))
```

```python
BLEU Score = BP * N-gram-precision
```
